```python
import pandas as pd
```

```python
loaded_df = pd.read_csv('test_data_with_predictions.csv')
result_df = loaded_df
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate metrics
accuracy = accuracy_score(result_df['Label'], result_df['PredictedLabel'])
precision = precision_score(result_df['Label'], result_df['PredictedLabel'])
recall = recall_score(result_df['Label'], result_df['PredictedLabel'])
f1 = f1_score(result_df['Label'], result_df['PredictedLabel'])

print(f"Accuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1 Score: {f1}")
```

```python
# Confusion matrix
cm = confusion_matrix(result_df['Label'], result_df['PredictedLabel'])

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```

```python
from sklearn.metrics import roc_auc_score, roc_curve
import matplotlib.pyplot as plt

roc_auc = roc_auc_score(result_df['Label'], result_df['PredictedLabel'])
print(f"ROC AUC Score: {roc_auc}")


```

```python
fpr, tpr, thresholds = roc_curve(result_df['Label'], result_df['PredictedLabel'])

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label=f'ROC curve (area = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
```
