```python
## Geting outliers fr 

def outliers(data, col, multiplier = 1.5):
    Q1 = data[col].quantile(0.25)
    Q2 = data[col].quantile(0.50)
    Q3 = data[col].quantile(0.75)
    Q4 = data[col].quantile(1)
    
    IQR = Q3 - Q1
    
    bas = Q1 - multiplier * IQR
    haut = Q3 + multiplier * IQR
    
    outliers = data[(data[col] < bas) | (data[col] > haut)]
    
    return outliers
```

```python
len(outliers(data,'Amount')) / len(data)

# Here we conclude that an outlier is not necessarily a fraud attitude
```

```python
# On va faire un box plot for outliers on amount ?
def box_plot(data, col):
    plt.figure(figsize=(12,8))
    plt.boxplot(data)
    plt.title(f'Box Plot of {col}')
    plt.show()
```

```python
def plot_scatterplot(data, column1, column2):
    plt.figure(figsize=(10,5))
    plt.scatter(data[column1], data[column2], alpha=0.2)
    plt.title(f'Scatter Plot of {column1} vs {column2}')
    plt.xlabel(column1)
    plt.ylabel(column2)
    plt.show()
```

```python
plot_scatterplot(data, 'Class', 'Amount')
```

```python
def plot_timeseries(data, time_column, class_column):
    plt.figure(figsize=(10,5))
    classes = data[class_column].unique()
    for i, cls in enumerate(classes):
        if i == 0:  
            alpha = 0.5  # Semi-transparent
        else:
            alpha = 1.0  # Fully opaque
        data[data[class_column] == cls].groupby(time_column).size().plot(kind='line', stacked=True, alpha=alpha)
    plt.title('Time Series Plot')
    plt.legend(['No fraud','Fraud'])
    plt.show()
```

```python
plot_timeseries(data, 'Time', 'Class')
```

```python
import matplotlib.pyplot as plt

class_counts = data['Class'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(class_counts, labels=class_counts.index, autopct='%1.1f%%', startangle=90, colors=['#66b3ff','#99ff99'])
plt.title('Distribution of Classes')
plt.legend(['No fraud','Fraud'])
plt.show()
```

```python
## let's use distribution plots of seaborn, we can use histplot but it looks ugly. 

df = data
fig, ax = plt.subplots(1, 2, figsize=(18,4))

amount_val = df['Amount'].values
time_val = df['Time'].values

sns.distplot(amount_val, ax=ax[0], color='r')
ax[0].set_title('Distribution of Transaction Amount', fontsize=14)
ax[0].set_xlim([min(amount_val), max(amount_val)])

sns.distplot(time_val, ax=ax[1], color='b')
ax[1].set_title('Distribution of Transaction Time', fontsize=14)
ax[1].set_xlim([min(time_val), max(time_val)])
```
